```python
import pandas as pd

data = {
    'ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen', 'Ivy', 'Jack'],
    'Age': [25, 30, 35, 40, 22, 29, 38, 27, 33, 31],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose'],
    'Salary': [50000, 60000, 75000, 80000, 45000, 62000, 72000, 58000, 69000, 71000],
    'Department': ['IT', 'HR', 'Finance', 'Marketing', 'IT', 'HR', 'Finance', 'Marketing', 'IT', 'Finance']
}

df = pd.DataFrame(data)

print(df)
```

取得 Name 和 Salary 二欄資料


```python
df[['Name','Salary']]
```

取出索引值 5 到 8 共 4 筆資料


```python
df.iloc[5:9]
```

找出年齡和薪水都高於平均值的人員

```python
avg_age = df["Age"].mean()
avg_salary = df['Salary'].mean()
print(f"平均年齡：{avg_age}")
print(f"平均薪水：{avg_salary}")
condition =  (df['Age'] > avg_age) & (df['Salary'] > avg_salary)
df[condition]
```

請問公司人員的平均薪水、及最高、最低薪水、中位數？

```python
print(f'公司人員的平均薪水：{df["Salary"].mean()}')
print(f'公司人員的最高薪水：{df["Salary"].max()}')
print(f'公司人員的最低薪水：{df["Salary"].min()}')
print(f'公司人員的中位數：{df["Salary"].median()}')
```

公司共有幾個部門？

```python
df['Department'].unique()
```

那個部門薪水最高？那個部門最年輕？

```python
print(df.groupby('Department')['Salary'].mean().idxmax(), df.groupby('Department')['Salary'].mean().max())
print(df.groupby('Department')["Age"].mean().idxmin())
```

年齡是否與薪資有關？

```python
corr_matrix = df[['Age', 'Salary']].corr()
corr_matrix
```
